```python
# ADVANCED: Iterate through all mappings in a workflow
def extract_all_mappings_from_workflow(workflow_json, workflow_name):
    """
    Extract all mappings from a workflow and add them to the Excel exporter
    """
    from new_style import InfaLineageGenerator
    
    # Create exporter for this workflow
    exporter = create_multi_mapping_exporter(f"{workflow_name}_all_mappings.xlsx")
    
    # Extract mappings from the workflow
    powermart = workflow_json.get('POWERMART', {})
    repository = powermart.get('REPOSITORY', {})
    folder_data = repository.get('FOLDER', {})
    
    # Handle FOLDER as either dict or list
    folders_to_process = []
    if isinstance(folder_data, dict):
        folders_to_process = [folder_data]
    elif isinstance(folder_data, list):
        folders_to_process = folder_data
    
    total_mappings = 0
    successful_mappings = 0
    
    for folder in folders_to_process:
        mapping_data = folder.get('MAPPING', [])
        
        # Handle both single mapping (dict) and multiple mappings (list)
        mappings = []
        if isinstance(mapping_data, dict):
            mappings = [mapping_data]
        elif isinstance(mapping_data, list):
            mappings = mapping_data
        
        for mapping in mappings:
            mapping_name = mapping.get('@NAME', 'Unknown')
            total_mappings += 1
            
            try:
                print(f"Processing mapping {total_mappings}: {mapping_name}")
                lineage_gen = InfaLineageGenerator(mapping_name, workflow_json)
                add_mapping_to_excel(exporter, lineage_gen)
                successful_mappings += 1
            except Exception as e:
                print(f"  ❌ Failed to process {mapping_name}: {str(e)}")
    
    # Finalize the Excel file
    finalize_multi_mapping_excel(exporter)
    
    print(f"\n📊 Summary:")
    print(f"   Total mappings found: {total_mappings}")
    print(f"   Successfully processed: {successful_mappings}")
    print(f"   Failed: {total_mappings - successful_mappings}")
    
    return exporter

# Example usage: Process all mappings from a workflow
print("Processing all mappings from WF_TRAN_STG workflow...")
exporter_tran_stg = extract_all_mappings_from_workflow(wf_tran_stg_json, "WF_TRAN_STG")
```

```python
# Process multiple workflows into a single Excel file
def process_multiple_workflows(workflow_dict, output_filename="all_workflows_comprehensive.xlsx"):
    """
    Process multiple workflows and combine all mappings into a single Excel file
    
    Args:
        workflow_dict: Dictionary with workflow names as keys and workflow JSON as values
        output_filename: Name of the output Excel file
    """
    # Create a single exporter for all workflows
    exporter = create_multi_mapping_exporter(output_filename)
    
    total_mappings = 0
    successful_mappings = 0
    
    for workflow_name, workflow_json in workflow_dict.items():
        print(f"\n🔄 Processing workflow: {workflow_name}")
        
        # Extract mappings from the workflow
        powermart = workflow_json.get('POWERMART', {})
        repository = powermart.get('REPOSITORY', {})
        folder_data = repository.get('FOLDER', {})
        
        # Handle FOLDER as either dict or list
        folders_to_process = []
        if isinstance(folder_data, dict):
            folders_to_process = [folder_data]
        elif isinstance(folder_data, list):
            folders_to_process = folder_data
        
        for folder in folders_to_process:
            mapping_data = folder.get('MAPPING', [])
            
            # Handle both single mapping (dict) and multiple mappings (list)
            mappings = []
            if isinstance(mapping_data, dict):
                mappings = [mapping_data]
            elif isinstance(mapping_data, list):
                mappings = mapping_data
            
            for mapping in mappings:
                mapping_name = mapping.get('@NAME', 'Unknown')
                total_mappings += 1
                
                try:
                    print(f"  📋 Processing: {mapping_name}")
                    lineage_gen = InfaLineageGenerator(mapping_name, workflow_json)
                    add_mapping_to_excel(exporter, lineage_gen)
                    successful_mappings += 1
                except Exception as e:
                    print(f"    ❌ Failed: {str(e)}")
    
    # Finalize the Excel file
    finalize_multi_mapping_excel(exporter)
    
    print(f"\n📊 Final Summary:")
    print(f"   Total mappings found: {total_mappings}")
    print(f"   Successfully processed: {successful_mappings}")
    print(f"   Failed: {total_mappings - successful_mappings}")
    print(f"   📁 Output file: {output_filename}")
    
    return exporter

# Example: Process all your workflows
all_workflows = {
    "WF_PLD_KYC2_UDM_STG_LOAD": wf_pld_kyc2_udm_stg_load_json,
    "WF_TRAN_STG": wf_tran_stg_json,
    "WF_STATIC_ACTZ": wf_static_actz_json,
    "WF_STATIC_STG": wf_static_stg_json,
    "WF_TRAN_ACTZ": wf_tran_actz_json,
    "WF_TRAN_PLD": wf_tran_pld_json,
    "WF_STATIC_PLD": wf_static_pld_json
}

# Uncomment the line below to process all workflows
# exporter_all = process_multiple_workflows(all_workflows, "ALL_WORKFLOWS_COMPREHENSIVE.xlsx")
```
